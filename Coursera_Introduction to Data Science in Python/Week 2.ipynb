# Querying a Series

```python
import pandas as pd
import numpy as np
```

```python
sports = {'Archery': 'Bhutan',
          'Golf': 'Scotland',
          'Sumo': 'Japan',
          'Taekwondo': 'South Korea'}
s = pd.Series(sports)
s
```

```python
s.iloc[3]
#Keep in mind that iloc and loc are not methods, they are attributes. So you don't use parentheses to query them, but square brackets instead
```

```python
s.loc['Golf']
# indexing operator
```

```python
s[3]
```

```python
s['Golf']
```

```python
sports = {99: 'Bhutan',
          100: 'Scotland',
          101: 'Japan',
          102: 'South Korea'}
s = pd.Series(sports)
```

```python
s[0] #This won't call s.iloc[0] as one might expect, it generates an error instead
# this creates confusion in python, use loc or iloc
```

```python
#iteration
s = pd.Series([100.00, 120.00, 101.00, 3.00])
s
```

```python
#this works but it is slow
total = 0
for item in s:
    total+=item
print(total)
```

```python
#numpy use vectorization to speed up the calculation
import numpy as np

total = np.sum(s)
print(total)
```

```python
#this creates a big series of random numbers
s = pd.Series(np.random.randint(0,1000,10000))
s.head()
# head will reduce the print out elements
```

```python
len(s)
```

```python
%%timeit -n 100
summary = 0
for item in s:
    summary+=item
#use % with [Tab] key to get a list of magic functions
#%% sign signal celler magic function and raptorcode
#%% must be on the first line otherwise error will thrown
```

```python
%%timeit -n 100
summary = np.sum(s)
#parallel computing feature is what data scienctist needs to know
```

```python
s+=2 #adds two to each item in s using broadcasting
s.head()
#broadcasting, apply an operation to every value in a series
```

```python
for label, value in s.iteritems():
    s.set_value(label, value+2)
s.head()
```

```python
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,10000))
for label, value in s.iteritems():
    s.loc[label]= value+2
#time the iteration approach
```

```python
%%timeit -n 10
s = pd.Series(np.random.randint(0,1000,10000))
s+=2
#time the parallel computing approach, it's significantly faster and more concise 
```

# Querying a DataFrame

```python
# boolean masking in numpy, its basically a True/False array/dataframe overlaying on top of the datafram to admit or not admit the final result
# boolean mask is aplied directly to the pandas object and is really powerful
df['Gold'] > 0
```

```python
only_gold = df.where(df['Gold'] > 0)#where function takes boolean mask to df and generate a new df
only_gold.head()
```

```python
only_gold['Gold'].count()#countries with no gold medals have values NaN, which are not counted here
```

```python
df['Gold'].count()
```

```python
only_gold = only_gold.dropna()#use dropna to drop NaN values, default drop is rows
only_gold.head()
```

```python
only_gold = df[df['Gold'] > 0] #more concise example
only_gold.head()
```

```python
len(df[(df['Gold'] > 0) | (df['Gold.1'] > 0)])#can combine boolean masks together to get more complex query
```

```python
df[(df['Gold.1'] > 0) & (df['Gold'] == 0)]#And operator. Each boolean mask needs to be in parenthesis
```
